```python
data = pd.read_csv('datafm.csv')
x = data["Calories"].mean()
data["Calories"].fillna(x, inplace = True)
print(data.to_string())
```

```python
data = pd.read_csv('datafm.csv')
x = data["Calories"].median()
data["Calories"].fillna(x, inplace = True)
print(data.to_string())
```
